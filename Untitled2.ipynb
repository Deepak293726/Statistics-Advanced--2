#Q1.What is hypothesis testing in statistics?
 =Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data. It involves proposing two hypotheses—the null hypothesis and the alternative hypothesis—and then using statistical evidence to determine whether to reject the null hypothesis. Hypothesis testing evaluates whether the observed data deviates significantly from what is expected under the null hypothesis, using a defined significance level (e.g., 0.05) to assess the strength of evidence.

#Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
 =The null hypothesis (H₀) is a statement that there is no effect or no difference, and it serves as the default or starting assumption in hypothesis testing. The alternative hypothesis (H₁) is the statement that contradicts the null hypothesis—it suggests that there is an effect or a difference. For example, if testing whether a new drug is effective, H₀ might state “the drug has no effect,” while H₁ would state “the drug has a significant effect.”


#Q3.Explain the significance level in hypothesis testing and its role in deciding the outcome of a test?
 =The significance level (α) is the probability threshold used to determine whether to reject the null hypothesis. Common values include 0.01, 0.05, and 0.10. If the p-value (probability of observing the data under H₀) is less than or equal to α, the null hypothesis is rejected, indicating statistically significant results. A smaller α means stronger evidence is required to reject H₀, helping control the risk of a Type I error (false positive).


#Q4.What are Type I and Type II errors? Give examples of each?
 =Type I Error: Rejecting the null hypothesis when it is actually true.
Example: Concluding a patient has a disease when they actually don’t.

Type II Error: Failing to reject the null hypothesis when it is false.
Example: Concluding a patient does not have a disease when they actually do.

#Q5.What is the difference between a Z-test and a T-test? Explain when to use each?
 =A Z-test is used when the population standard deviation is known and the sample size is large (n ≥ 30).

A T-test is used when the population standard deviation is unknown and the sample size is small (n < 30).

Use Z-test: Comparing a sample mean to a known population mean with known σ.

Use T-test: When σ is unknown and the sample is small, such as comparing two small group means.

```python
#Q6.Write a Python program to generate a binomial distribution with n=10 andp=0.5, then plot its histogram?
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 1000

# Generate binomial data
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(n+2)-0.5, edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

```

```python
#Q7.Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results?
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Convert to array
data = np.array(sample_data)
sample_mean = np.mean(data)
population_mean = 50
sample_std = np.std(data, ddof=1)
n = len(data)

# Z-statistic
z = (sample_mean - population_mean) / (sample_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Z-statistic:", z)
print("P-value:", p_value)
```

```python
#Q8.Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib?
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate data
data = np.random.normal(loc=50, scale=5, size=100)

# Mean and standard error
mean = np.mean(data)
sem = stats.sem(data)

# Confidence Interval
conf_interval = stats.t.interval(0.95, df=len(data)-1, loc=mean, scale=sem)

print("95% Confidence Interval:", conf_interval)

# Plot data
plt.hist(data, bins=15, edgecolor='black')
plt.axvline(conf_interval[0], color='red', linestyle='dashed', label='Lower CI')
plt.axvline(conf_interval[1], color='green', linestyle='dashed', label='Upper CI')
plt.title("Normal Distribution Sample")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()

```

```python
#Q9.Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores representin terms of standard deviations from the mean?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Sample data
data = np.random.normal(50, 5, 100)

# Calculate Z-scores
z_scores = zscore(data)

# Plot histogram
plt.hist(z_scores, bins=15, edgecolor='black')
plt.title("Z-score Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# Explanation
print("Z-scores represent how many standard deviations each value is from the mean.")

```
